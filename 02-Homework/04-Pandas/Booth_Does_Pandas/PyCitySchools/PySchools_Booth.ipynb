# Import Data

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

# District Summary

```python
#create columns
totSchools = df.school_name.nunique()
totStudents = df["Student ID"].nunique()
totBudget = df.groupby("school_name").budget.mean().sum()
avgMath = df.math_score.mean()
avgReading = df.reading_score.mean()

mask1 = df.math_score >= 70
percMath = 100 * df.loc[mask1, ["Student ID"]].count().values[0] / df["Student ID"].nunique()

mask2 = df.reading_score >= 70
percReading = 100 * df.loc[mask2, ["Student ID"]].count().values[0] / df["Student ID"].nunique()

mask3 = (df.math_score >= 70) & (df.reading_score >= 70)
percBoth = 100 * df.loc[mask3, ["Student ID"]].count().values[0] / df["Student ID"].nunique()

#create table
districtSummary = pd.DataFrame()
districtSummary["Total Schools"] = [totSchools]
districtSummary["Total Students"] = [totStudents]
districtSummary["Total Budget"] = [totBudget]
districtSummary["Average Math Score"] = [avgMath]
districtSummary["Average Reading Score"] = [avgReading]
districtSummary["% Passing Math"] = [percMath]
districtSummary["% Reading Score"] = [percReading]
districtSummary["% Overall Passing"] = [percBoth]

#print
districtSummary
```

# School Summary

```python
school_groups = df.groupby("school_name")

#get columns
tot_students = school_groups.size()
types = school_groups.type.unique()
budget = school_groups.budget.mean()
budgetPerStudent = budget / tot_students
math = school_groups.math_score.mean()
reading = school_groups.reading_score.mean()
mathperc = df.loc[df.math_score >= 70].groupby("school_name").size() / tot_students
readingperc = df.loc[df.reading_score >= 70].groupby("school_name").size() / tot_students
totalperc = df.loc[(df.math_score >= 70) & (df.reading_score >= 70)].groupby("school_name").size() / tot_students

#Build the table
schoolSummary = pd.DataFrame()
schoolSummary["School Type"] = types
schoolSummary["School Type"] = [x[0] for x in schoolSummary["School Type"]] #hack because idk how to code
schoolSummary["Total Students"] = tot_students
schoolSummary["Total School Budget"] = budget
schoolSummary["Per Student Budget"] = budgetPerStudent
schoolSummary["Average Math Score"] = math
schoolSummary["Average Reading Score"] = reading
schoolSummary["% Passing Math"] = mathperc
schoolSummary["% Passing Reading"] = readingperc
schoolSummary["% Overall Passing"] = totalperc


#print the table
schoolSummary
```

# Math Scores by Grade

```python
# loop through each grade, filter, groupby school, and save the math score series
math_grades = []

for grade in df.grade.unique():
    df_sub = df.loc[df.grade == grade]
    temp = df_sub.groupby("school_name").math_score.mean()
    math_grades.append(temp)

#concat all the series together and sort out the columns
mathSummary = pd.concat(math_grades, axis=1)
mathSummary.columns = df.grade.unique()
mathSummary = mathSummary[["9th", "10th", "11th", "12th"]]
mathSummary
```

# Reading Score by Grade

```python
# loop through each grade, filter, groupby school, and save the math score series
reading_grades = []

for grade in df.grade.unique():
    df_sub = df.loc[df.grade == grade]
    temp = df_sub.groupby("school_name").reading_score.mean()
    reading_grades.append(temp)

#concat all the series together and sort out the columns
readingSummary = pd.concat(reading_grades, axis=1)
readingSummary.columns = df.grade.unique()
readingSummary = readingSummary[["9th", "10th", "11th", "12th"]]
readingSummary
```
